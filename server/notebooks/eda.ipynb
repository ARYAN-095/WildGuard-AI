```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Load metadata CSV (adjusted for notebook location)
meta = pd.read_csv("../data/metadata/Metadata V1.0 FSC22.csv")

# Check the column names to use correct keys
print(meta.columns.tolist())  # This will help verify the column names

# Use the actual column name from the CSV
# Correct column name is 'Dataset File Name' and 'Class Name'
meta["filepath"] = meta["Dataset File Name"].apply(
    lambda fn: os.path.join("data/raw/Audio Wise V1.0", fn + ".wav")
)

# Plot class distribution (use 'Class Name' column)
meta["Class Name"].value_counts().plot.barh(figsize=(10, 6), title="Class Distribution")
plt.xlabel("Number of Samples")
plt.tight_layout()
plt.show()
```
